```python
# 재귀함수recursive function 구현 예
def factorial_for(n):
    result = 1
    for item in range(1, n + 1, 1):
        result *= item
    return result


def factorial_recr(n):
    if n > 1:
        return n * factorial_recr(n - 1)
    else:
        return 1
```

```python
factorial_for(10)
```

```python
factorial_recr(10)
```

```python
factorial_for(100)
```

```python
factorial_recr(100)
```

```python
factorial_recr(1000)
```

```python
factorial_recr(100000)
```

```python
def hello():
    print('hello 함수 시작')
    print('hello')
    print('hello 함수 끝')

def world():
    print('world 함수 시작')
    print('world')
    print('world 함수 끝')

hello()
world()
```

```python
def trace(func):                             # 호출할 함수를 매개변수로 받음
    def wrapper():                           # 호출할 함수를 감싸는 함수
        print('함수 시작')    # __name__으로 함수 이름 출력
        func()                               # 매개변수로 받은 함수를 호출
        print('함수 끝')
    return wrapper                           # wrapper 함수 반환


def hello():
    print("hello")

trace_hello = trace(hello)    # 데코레이터에 호출할 함수를 넣음
trace_hello()                 # 반환된 함수를 호출
```

```python
@trace    # @데코레이터
def hello():
    print('hello')

hello()    # 함수를 그대로 호출
```

```python
df_d = pd.get_dummies(df, columns=["color", "carrier", "noDescription", "productline"], drop_first=True)
```

```python
df_d.head()
```

```python
df["sold"] = pd.Series(df["sold"])
```

```python
X = df_d.drop("sold", axis=1)
y = df_d["sold"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth=3)
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
pred[:5]
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, pred)
```

```python
confusion_matrix(y_test, pred)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
```

```python
model_list = [LogisticRegression, KNeighborsClassifier, GradientBoostingClassifier, RandomForestClassifier]
```

```python
for i in model_list:
    model = i()
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    print(model.__str__(), accuracy_score(y_test, pred))
```

```python
test_size = [0.05, 0.10, 0.15, 0.20]
max_depth = [2,3,4,5,6,7,8,9,10]

for ts in test_size:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=ts, stratify=df_d["sold"])
    for md in max_depth:
        model = DecisionTreeClassifier(max_depth=md)
        model.fit(X_train, y_train)
        pred = model.predict(X_test)
        print(f"test size: {ts} max_depth: {md} score: {accuracy_score(y_test, pred)}")
```
