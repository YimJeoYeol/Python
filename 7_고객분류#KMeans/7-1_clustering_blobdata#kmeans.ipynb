# 모듈 로딩 및 랜덤 데이터 생성

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples = 200, n_features = 2, centers = 4, random_state = 100)
```

```python
# 해당 데이터는 독립변수와 종속변수가 각각 별도의 Numpy Array로 존재하여, data라는 이름의 Numpy Array에 묶여있음
data
```

```python
# 독립변수: 0으로 인덱싱
data[0]
```

```python
# 종속변수: 1로 인덱싱
data[1]
```

```python
# 독립변수에서 첫번째 컬럼을 인덱싱하는 방법
data[0][:, 0]
```

```python
sns.scatterplot(x = data[0][:, 0], y = data[0][:, 1])
```

# KMeans Clustering Modeling

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 4)
```

```python
model.fit(data[0])
```

```python
# 학습된 모델의 클러스터명 확인
model.labels_
```

```python
data[1]
```

```python
# 실제값으로 그려본 scatter plot
sns.scatterplot(x = data[0][:, 0], y = data[0][:, 1], hue = data[1], palette='Set2')
```

```python
# 모델링 결과로 그려본 scatter plot
sns.scatterplot(x = data[0][:, 0], y = data[0][:, 1], hue = model.labels_, palette='Set2')
```

### 다른 k값으로 클러스터링 해보기

```python
model = KMeans(n_clusters = 3)
```

```python
model.fit(data[0])
```

```python
sns.scatterplot(x = data[0][:, 0], y = data[0][:, 1], hue = model.labels_, palette='Set2')
```

```python
model = KMeans(n_clusters = 5)
model.fit(data[0])
sns.scatterplot(x = data[0][:, 0], y = data[0][:, 1], hue = model.labels_, palette='Set2')
```

# 최적의 K값 찾기: Elbow method

```python
model.inertia_
```

```python
distance = []

for i in range(2, 11):
    model = KMeans(n_clusters = i)
    model.fit(data[0])
    distance.append(model.inertia_)
    
```

```python
distance
```

```python
list(range(2,11))
```

```python
sns.lineplot(x=list(range(2,11)), y = distance)
```

### Plot에서 급격히 꺾이는 부분이 최적의 k값
